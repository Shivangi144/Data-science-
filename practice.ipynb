```python
# Inheritance
class Address:
  def __init__(self,city,state):
    self.city = city
    self.state = state
  def location(self):
    return f"my location is {self.city} in {self.state}"
class User(Address):
  def __init__(self,name,age,city,state):
    self.name = name
    self.age = age
    self.city = city
    self.state= state
  def userName(self):
    print(f"My name is {self.name} and my location is {self.city} , {self.state}")
    print(f"My age is {self.age}")
u = User("Shivangi",20,"Jaipur","Rajasthan")
print(u.city)
print(u.state)
u.userName()

a = Address("Jaipur","Rajasthan")




```
