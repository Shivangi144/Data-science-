Line chart

```python
import matplotlib.pyplot as plt # visualization
x = [2005,2006,2007,2008]
y = [200,300,450,500]
plt.plot(x,y) # line graph
plt.xlabel("years")
plt.ylabel("sales data")
plt.show()

# graph size
plt.figure(figsize=(6,2))
plt.plot(x,y,color="blue",marker='o',linestyle="dashed",linewidth=5,markersize=16)
plt.show()

# multi lines chart
import matplotlib.pyplot as plt
x = [2005,2006,2007,2008]
y1 = [200,300,450,500]
y2 = [150,350,500,550]
plt.plot(x,y)
plt.plot(x,y1,label="tops",color="red",linewidth=4)
plt.plot(x,y2,label="skirts",color="blue",linewidth=2)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
months = ["Jan","Feb","Mar","Apr"]
sales1 = [150,250,300,400]
sales2 = [200,300,450,550]
plt.plot(months,sales1)
plt.xlabel("months")
plt.ylabel("sales data")
plt.title("Monthly Sales Data")
plt.plot(months,sales1,label="Novels",marker='o',color ="green",linewidth=2,markersize=12)
plt.plot(months,sales2,label="Self help books",marker='v',color="red",linewidth=3,markersize=12)
plt.legend()
plt.show()
```

Bar chart

```python
# Bar chart
import matplotlib.pyplot as plt
x = [2000,2005,2010,2015]
y = [100,150,250,400]
plt.bar(x,y)
# size
plt.figure(figsize=(6,2))
plt.show()

# multi bar chart
import matplotlib.pyplot as plt
import numpy as np
months =["Jul","Aug","Sept","Oct"]
sales1 = [200,350,400,450]
sales2 = [300,400,450,550]
x = np.arange(len(months))
w= 0.40
plt.xlabel("months")
plt.ylabel("sales")
plt.bar(x - w/2,sales1,label="Dairy Milk",color="green",width=w)
plt.bar(x + w/2,sales2,label="Kit Kat",color="blue",width=w)
plt.xticks(x, months)
plt.legend()
plt.show()
#
import matplotlib.pyplot as plt
import numpy as np
months= ["Jun","Jul","Aug","Sept","Oct"]
y1 = [100,150,250,350,400]
y2 = [145,200,280,350,450]
y3 =[150,225,350,400,500]
x = np.arange(len(months))
w = 0.40
plt.xlabel("months")
plt.ylabel("sales")
plt.bar(x-w,y1,width=w,color="green",label="Waffles")
plt.bar(x,y2,width=w,color="blue",label="Jar cake")
plt.bar(x+w,y3,width=w,color="red",label="Donuts")
plt.title("Sales of bakery")
plt.xticks(x, months)
plt.legend()
plt.show()



```

Histogram & Pie Chart


```python
# Histogram
import matplotlib.pyplot as plt
import numpy as np
marks =[40,55,60,70,90,75,33,50]
plt.hist(marks,bins = 5,color = "green")
plt.show()

# pie chart
import matplotlib.pyplot as plt
fruits = ['apple','strawberry','litchi','mango']
count = [40,30,45,70]
plt.pie(count, labels=fruits)
plt.show()

#pie chart percentage
import matplotlib.pyplot as plt
fruits = ['strawberry','litchi','mango','cherry']
colors = ['blue','red','yellow','green']
count = [30,60,85,75]
plt.pie(count,labels=fruits,colors=colors)
plt.show()
```

Subplots

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1000,1500,2500,3500,2000]
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.subplot(1,2,1)
#
x1 = ['litchi','mango','muskmelon','cherry']
y1 = [50,75,45,90]
plt.pie(y1,labels=x1)
plt.xlabel("x1 axis")
plt.ylabel("y1 axis")
plt.subplot(1,2,2)
plt.show()

# sub plot
import matplotlib.pyplot as plt
# first chart
x = [1,2,3,4,5]
y = [100,150,300,225,400]

plt.subplot(2,2,1) # row,cols,position
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")

# second chart
x1 = ['apple','banana','mango','litchi']
y1 = [40,30,15,70]

plt.subplot(2,2,2) # row,cols, position
plt.pie(y1,labels=x1,startangle=90)
plt.xlabel("x1 axis")
plt.ylabel("y1 axis")

# third chart
x = [1,2,3,4,5]
y = [100,180,250,300,200]

plt.subplot(2,2,3) # row,cols,position
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
# fourth chart
x1 = ['apple','banana','orange','watermelon']
y1 = [40,30,15,70]

plt.subplot(2,2,4) # row,cols, position
plt.pie(y1,labels=x1,startangle=90)
plt.xlabel("x1 axis")
plt.ylabel("y1 axis")


plt.tight_layout()
plt.show()
```
